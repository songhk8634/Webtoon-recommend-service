```python
from konlpy.tag import Mecab
mec = Mecab(dicpath = 'c:/mecab')
```

```python
review = pd.read_csv('review_table.csv')
review2 = review[['review', 'title']]

# 댓글 한글만 추출
review2['review'] = review2['review'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")
title = review2['title'].unique().tolist()    ## 제목 리스트


## 작품별 댓글 통합

sentence = []
k = ''
count = 0
for i in range(len(title)):                       # 제목 갯수
    for j in range(len(review2)):                          # 댓글 갯수
        if review2['title'][j] == title[i]:
            k += review['review'][j]
            
            count += 1
    if k == '':
        pass
    else :
        sentence.append(k)
    k = ''
    
    print(count / len(review2) * 100)
```

```python
review_tot = pd.DataFrame()
review_tot['review'] = sentence
review_tot['title'] = title
review_tot['review'] = review_tot['review'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")

review_tot['review2'] = review_tot['review'].str.replace("[^가-힣 ]","")  ## ㅋㅋ제거

review_tot2 = pd.DataFrame()
review_tot2['review'] = review_tot['review2']
review_tot2['title'] = title
```

```python
sep = []

for word in review_tot2['review'] :
    sep.append(mec.pos(word))
```

```python
############################# keyword 

keyword = []
blank = []
for word in sep:
    for each in word:
        if each[1] == 'NNG':          # 명사만 뽑기
            blank.append(each[0])
        elif each[1] == 'NNP':
            blank.append(each[0])
        else:
            pass
    keyword.append(blank)
    blank = []
    
keyword2 = []                      ### 두글자 이상만 뽑기
blank = []
for word in keyword:
    for each in word:
        if len(each) >= 2 :
            blank.append(each)
    keyword2.append(blank)
    blank = []

review_tot2['keyword'] = keyword2
```

```python
####################################### TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer
from collections import defaultdict
from operator import itemgetter

tf_word = []                                          ## 모든 keyword2 이어 붙이기
for i in range(len(review_tot2['keyword'])):
    tf_word.append(' '.join(t for t in review_tot2['keyword'][i]))


tfidf = TfidfVectorizer()
sp_matrix = tfidf.fit_transform(tf_word)

word2id = defaultdict(lambda : 0)
for idx, feature in enumerate(tfidf.get_feature_names()):
    word2id[feature] = idx

word = []
for i, sent in enumerate(tf_word):
    word.append([ [token, sp_matrix[i, word2id[token]]] for token in sent.split()])

for i in range(len(word)):
    word[i].sort(key = itemgetter(1), reverse = True)      # 중요도 순으로 정렬

print('tf-idf 계산 완료')
```

```python
##########################        한 작품에 여러 단어가 들어가 있으므로 중복 제거
new_word = []
blank = []

for i in range(len(word)):
    
    for j in range(len(word[i])):
        if word[i][j] not in blank :
            blank.append(word[i][j])
        else :
            pass
    new_word.append(blank)  
    blank = []
    ################################              상위 10개 뽑기
hot_word = []
blank = []
for i in range(len(new_word)):
    for j in range(len(new_word[i])):
        blank.append(new_word[i][j][0])
    hot_word.append(blank[:10])
    blank = []

review_tot2['hot_word'] = hot_word
```

```python
sep = []

for word in summary['summary'] :
    sep.append(mec.pos(word))
```

```python
############################# keyword 

keyword = []
blank = []
for word in sep:
    for each in word:
        if each[1] == 'NNG':          # 명사만 뽑기
            blank.append(each[0])
        elif each[1] == 'NNP':
            blank.append(each[0])
        else:
            pass
    keyword.append(blank)
    blank = []
    
keyword2 = []                      ### 두글자 이상만 뽑기
blank = []
for word in keyword:
    for each in word:
        if len(each) >= 2 :
            blank.append(each)
    keyword2.append(blank)
    blank = []

summary['keyword'] = keyword2
```

```python
## sentence vector 얻는 함수
## 문장의 벡터는 해당 문장의 단어들의 벡터평균

def get_sentence_mean_vector(morphs):
    vector = []
    for i in morphs:
        try:
            vector.append(model.wv[i])
        except KeyError as e:
            pass
    try:
        return np.mean(vector, axis = 0)
    except IndexError as e:
        pass
```

```python
dfsum['wv'] = dfsum['keyword'].map(get_sentence_mean_vector)
```

```python
import matplotlib.pyplot as plt

## KMeans 사용
from sklearn.cluster import KMeans
import time

word_vectors = dfsum.wv.to_list()
```

```python
# 엘보우 기법 (오차제곱합의 값이 최소가 되도록 결정하는 방법)
def elbow(data, length):
    sse = [] # sum of squre error 오차제곱합
    for i in range(1, length):
        kmeans = KMeans(n_clusters=i)
        kmeans.fit(data)
        # SSE 값 저장
        sse.append(kmeans.inertia_)
    plt.plot(range(1, length), sse, 'bo-')
    plt.title("elbow method")
    plt.xlabel("number of clusters")
    plt.ylabel("SSE")
    plt.show()
elbow(word_vectors, 20)
```

```python
def elbow(data, length):
    from sklearn.metrics import silhouette_score
    sil_avg = [] # sum of squre error 오차제곱합
    plt.figure(figsize=(20,10))
    for i in range(2, length):
        kmeans = KMeans(n_clusters=i)
        kmeans.fit(data)
        # SSE 값 저장
        sil_avg.append(silhouette_score(data, kmeans.fit_predict(data)))
    plt.plot(range(2, length), sil_avg, 'bo-')
    plt.title("silhouette method")
    plt.xlabel("number of clusters")
    plt.ylabel("sil_avg")
    plt.show()
elbow(word_vectors, 20)
```

```python
## K means 를 정의하고 학습
num = 6
kmeans_clustering = KMeans(n_clusters = num)
idx = kmeans_clustering.fit_predict(word_vectors)
dfsum['category'] = idx
```

```python
## 차원축소 위한 TSNE

from sklearn.manifold import TSNE

X = dfsum['wv'].to_list()
y = dfsum['category'].to_list()

import os.path
import pickle

tsne_filepath = 'tsne300.pkl'
#########################################
# File Cache

if not os.path.exists(tsne_filepath):
    tsne = TSNE(random_state = 42)
    tsne_points = tsne.fit_transform(X)
    with open(tsne_filepath, 'wb+') as f:
        pickle.dump(tsne_points, f)
else: # Cache Hits!
    with open(tsne_filepath, 'rb') as f:
        tsne_points = pickle.load(f)

tsne_df = pd.DataFrame(tsne_points, index = range(len(X)), columns = ['x_coord', 'y_coord'])
tsne_df['title'] = dfsum['name'].to_list()
tsne_df['summ'] = dfsum['summary'].to_list()
tsne_df['cluster_no'] = y


## Bokeh로 2차원 plotting

from bokeh.plotting import figure, show, output_notebook
from bokeh.models import HoverTool, ColumnDataSource, value
from bokeh.palettes import brewer

output_notebook()

#########################################
## Get the number of colors we'll need for the plot

colors = brewer['Spectral'][len(tsne_df['cluster_no'].unique())]

# Create a map between factor and color
colormap = {i : colors[i] for i in tsne_df['cluster_no'].unique()}

# Create a list of colors for each value that we will be looking at
colors = [colormap[x] for x in tsne_df['cluster_no']]

tsne_df['color'] = colors

#########################################
## Bokeh Datasource 만들기

plot_data = ColumnDataSource(
    data = tsne_df.to_dict(orient = 'list')
)

#########################################

# Plot 만들기(배경)

tsne_plot = figure(
    title = 'TSNE Webtoon Summary Embeddings', 
    plot_width = 600,
    plot_height = 450,
    active_scroll = 'wheel_zoom',
    output_backend = 'webgl'         # GPU가속을 통해 부드러운 차트
)
#########################################

# 해당 Hover 튤립 만들기

tsne_plot.add_tools(
    HoverTool(
        tooltips = [('cluster', '@cluster_no'),('title','@title')]
    )
)
#########################################

## 해당 plot 에 데이터 정보들을 넣어준다.
## Source를 입력, x,y축, color정보를 컬럼명으로 넣어준다.

tsne_plot.circle(
    source = plot_data,
    x = 'x_coord',
    y = 'y_coord',
    line_alpha = 0.3,
    fill_alpha = 0.2,
    size = 10,
    fill_color = 'color',
    line_color = 'color',
    
)
#########################################
## 각 값들 추가

tsne_plot.title.text_font_size = value('16pt')
tsne_plot.xaxis.visible = False
tsne_plot.yaxis.visible = False
tsne_plot.grid.grid_line_color = None
tsne_plot.outline_line_color = None

show(tsne_plot)
```

# ========================  추천  
tfidf 유사도

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

```python
basic = pd.read_csv('basic_table.csv')
```

```python
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(basic['summary'])
print(tfidf_matrix.shape)
```

```python
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
indices = pd.Series(basic.index, index = basic['name']).drop_duplicates()
print(indices.head())
```

```python
def get_recommend(title, cosine_sim = cosine_sim):
    # 선택한 웹툰의 타이틀로부터 인덱스 추출
    idx = indices[title]
    
    # 모든 웹툰에 대해서 해당 웹툰과 유사도를 구함
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    # 유사도에 따라 웹툰 정렬
    sim_scores = sorted(sim_scores, key = lambda x: x[1], reverse = True)
    
    # 가장 유사한 10개 웹툰
    sim_scores = sim_scores[1:11]
    
    # 가장 유사한 10개 웹툰의 인덱스 추출
    webtoon_indices = [i[0] for i in sim_scores]
    
    # 가장 유사한 10개 웹툰의 제목
    return basic[['name', 'summary', 'genre']].iloc[webtoon_indices]
```
